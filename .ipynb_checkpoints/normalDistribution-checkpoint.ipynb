```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mu, sigma, n = 0,5,40
np.random.seed(12)# So that graphics are the same at next call
data = np.random.normal(mu,sigma, n)
f, ax = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)}, figsize=(10,6))
sns.boxplot(data, ax=ax[0])
sns.distplot(data, ax=ax[1], hist = True)
output =  plt.suptitle("Normal distribution boxplot and histogram")
output = ax[0].set(xlabel='')
ax[0].set_xticks(range(-20,22, 2))
output = ax[1].set(xlabel = 'Value', ylabel = "Frequency")
```

### Boxplot divided data into 4 quartiles and identified one data point as an outlier.
IQR (Interquartile range is used to identify an outlier
## TODO! Check how percentiles are calculated

```python
p75, p25 = np.percentile(data,75), np.percentile(data,25)
iqr = p75-p25
upperLimit, lowerLimit = p75+(iqr*1.5), p25-(iqr*1.5)
print(round(lowerLimit,2)," <- Outlier Limits -> ",round(upperLimit,2) )
```

Outlier may indicate an error in data (then it could be deleted) or unusual event which was missed during study and data collection.

## Normal Distrubition Formulas
$$
\text{Mean of Population: } \mu \\
\text{ observations number: n} \\
\text{Standard deviation: } \sigma  = \sqrt{\frac{\sum\limits_{i=1}^n(X_{i}-\mu)^2}{n}}\\
\text{Variance: } \sigma^2\\
\text{Data point: } x\\
\text{Gaussian/Normal distribution: } N(\mu, \sigma) \\
\text{Probability density: } P = \phi \\
\text{Z value: } z = \frac{x-\mu}{\sigma} \\ 
\text{Probability density function: } \phi = \frac{1}{\sqrt{2 \pi}\sigma}e^\frac{-\sqrt{x-\mu^2}}{2\mu^2}\\
$$

## Three sigma rule:
$$
\mu-1\sigma \leq X \leq \mu+1\sigma \approx 68.27\% \text{ of data} \\
\mu-2\sigma \leq X \leq \mu+2\sigma \approx 95.45\% \text{ of data} \\
\mu-3\sigma \leq X \leq \mu+3\sigma \approx 99.73\% \text{ of data} \\
$$
